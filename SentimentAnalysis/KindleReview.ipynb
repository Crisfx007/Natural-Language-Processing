```python
import pandas as pd
data = pd.read_csv('all_kindle_review.csv')
data.head()
```

```python
df = data[['reviewText','rating']]
```

```python
### Preprocessing and Cleaning.
### Positive Review is 1 and Negative Review is 0
df['rating']=df['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
df['rating']
```

```python
df.rating.value_counts()
```

```python
## Convert into lowercase
df['reviewText']=df['reviewText'].str.lower()
```

```python
df.head()
```

```python
### Cleaning the Data
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from bs4 import BeautifulSoup
```

```python
### Clean the Special Characters using Regex
df['reviewText'] = df['reviewText'].apply(lambda x:re.sub(r'[^a-zA-Z0-9\s]','',x))
```

```python
### Removing Stopwords.
sw_list = stopwords.words('english')
```

```python
df['reviewText']= df['reviewText'].apply(lambda x:[item for item in x.split() if item not in sw_list]).apply(lambda x:' '.join(x))
```

```python
### Remove URLs
df['reviewText']=df['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
```

```python
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from nltk.stem import WordNetLemmatizer
lm = WordNetLemmatizer()
```

```python
def lemmatize(text):
  return " ".join([lm.lemmatize(word) for word in text.split()])
```

```python
nltk.download('wordnet')
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmatize(x))
```

```python
df.head()
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df['reviewText'],df['rating'],test_size=0.2,random_state=42)
```

```python
## Using BOW
from sklearn.feature_extraction.text import CountVectorizer
bow = CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
#using TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
X_train_tfidf=tfidf.fit_transform(X_train).toarray()
X_test_tfidf=tfidf.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow=GaussianNB().fit(X_train_bow,y_train)
nb_model_tfidf=GaussianNB().fit(X_train_tfidf,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
y_pred_bow=nb_model_bow.predict(X_test_bow)
y_pred_tfidf=nb_model_bow.predict(X_test_tfidf)
```

```python
print("BOW accuracy: ",accuracy_score(y_test,y_pred_bow))
print("TFIDF accuracy",accuracy_score(y_test,y_pred_tfidf))
```

```python
confusion_matrix(y_test,y_pred_tfidf)
```

```python
confusion_matrix(y_test,y_pred_bow)
```
