```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.txt',
                    sep='\t',names=["label","message"])
```

```python
messages
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
import nltk
nltk.download('stopwords')
```

```python
corpus = []

for i in range(0,len(messages)):
    review = re.sub('[^a-zA]',' ',messages['message'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review =' '.join(review)
    corpus.append(review)
```

```python
corpus
```

### Using BOW

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
X
```

```python
cv.vocabulary_
```

```python
y=pd.get_dummies(messages['label'])
```

```python
y=y.iloc[:,0].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### Using TFIDF

```python
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
spam_tfidf_model = MultinomialNB().fit(X_train, y_train)
```

```python
y_pred=spam_tfidf_model.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```

```python
### We should always do train test split before doing any vectorization like we did in TFIDF.
### If we don't do it then the test data will not be unseen and impact the model with unusual accuracy and won't be good for unseen data.
```
