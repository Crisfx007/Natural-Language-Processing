```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
!git clone https://huggingface.co/fse/word2vec-google-news-300
!cd word2vec-google-news-300
```

```python
model = KeyedVectors.load("/content/word2vec-google-news-300/word2vec-google-news-300.model")
```

```python
model['man']
```

```python
model['man'].shape
```

```python
model.most_similar('man')
```

```python
model.most_similar('cricket')
```

```python
model.similarity('man','woman') ## checks cosine similarity between the two words by converting them into vectors
```

```python
model.similarity('man','PHP')
```

```python
model.doesnt_match(['PHP','java','monkey']) ## finds the odd one out
```

```python
## Vector arithmetic
vec = model['king'] - model['man'] + model['woman']
vec
```

```python
model.most_similar(vec)
```

```python
##
model.most_similar(positive=['king','woman'],negative=['man'])
```

```python
model.most_similar(positive=['INR','England'],negative=['India'])
```

```python
vec1 = model['INR'] - model['India'] + model['England']
vec1
```

```python
model.most_similar(vec1)
```

```python
# Manual vector arithmetic (e.g. king - man + woman) treats each vector equally and directly, 
# while most_similar(positive=['king', 'woman'], negative=['man'])
#internally averages the positive vectors before subtracting the negatives. This subtle difference can lead to different results.
#Which is better?
# The positive/negative method is preferred — it’s more stable, normalized, and accounts for multiple words naturally. 
# Use manual arithmetic only when you need full control over the math.
```
